# Binary Tree Inorder Traversal Iterative

Given a binary tree, return the inorder traversal of its nodes' values.

Example:
```
Input: [1,2,4,5,3]  

            1
          /   \
        2      3
      /  \
    4     5

Output: [4,2,5,1,3]
```

Here we are going to traverse through the Binary Tree Inorder Traversal Iterative:

We are going to use the stack in order to go back to the parent node or root node of the element.

* ### Description:

    * Given the Binary tree, create a variable named as **current** to store the **current root node**.
    * Create a stack to store current nodes which are not empty, this stack will help in traversing back to the root node.

* ### Given the above Binary Tree:
    * **Step 1:** Set the current address of the root to current. i.e current -> 1 
    <br/>
    
    * **Step 2:** Push the current node on the stack and set the current = current-> left until the current is None or Null.
       * current -> 1
       * Push 1: Stack [1]
            * set the current, current = current -> left i.e current -> 2
       * Push 2: Stack [1,2]
           * set the current, current = current-> left i.e current -> 4
       * Push 4: Stack [1,2,4]
           * set the current, current = current-> left i.e current -> None
       * ***Reached end of the left sub tree***  
       
       <br/>
     
    * **Step 3:** As the current is None. We will jump to the stack and pop the last element of the stack i.e 4.
        * set the last item of stack to current. i.e current = stack.pop()
        * Pop 4: Stack [1,2]
        * print 4
        * set the current = current -> right
            * As the right of the current is None. Again Step 3 is executed.
            
         <br/>
     
    * **Step 3:** Repeat the steps:
        * Pop 2: Stack [1]
        * print 2
        * set the current =  current -> right.
            * Here the current is 2 and 2 has a right child i.e 5. Thus current  = 5. ***Push the 5 to stack, go to step 2:***
         
         <br/>    
     
    * **Step 2:** Push the current node to stack.
        * Push 5: Stack [1,5]
            * set the current, current = current->left i.e current -> None
            * ***Reached end of the right sub tree***
         
         <br/>       
     
    * **Step 3:** Jump to the stack and pop the last element of the stack i.e 5.
        * set the last item of stack to current. i.e current = stack.pop()
        * Pop 5: Stack [1]
        * print 5
        * set the current = current -> right
            * As the right of the current is None. Again Step 3 is executed.
         
         <br/>       
     
    * **Step 3:** Jump to the stack and pop the last element of the stack i.e 1.
        * set the last item of stack to current. i.e current = stack.pop()
        * Pop 1: Stack []
        * print 1
        * set the current = current -> right
            * Here the current is 1 and 1 has a right child i.e 3. Thus current = 3. ***Push the 3 to stack, go to step 2:***
         
         <br/>
     
    * **Step 2:** Push the current node to stack.
        * Push 3: Stack [3]
            * set the current, current = current->left i.e current -> None
            * ***Reached end of the right sub tree***
         
         <br/>   
     
    * **Step 3:** Jump to the stack and pop the last element of the stack i.e 3.
        * set the last item of stack to current. i.e current = stack.pop()
        * Pop 3: Stack []
        * print 3
        * set the current = current -> right
           * ***Reached end of the right sub tree***
    * ***As the Current is None and Stack is also Empty. The else part will be executed and the function will break***
            
     
        
                  
    


````python
# Python program to do inorder traversal without recursion 
  
# A binary tree node 
class Node: 
      
    # Constructor to create a new node 
    def __init__(self, data): 
        self.data = data  
        self.left = None
        self.right = None
  
# Iterative function for inorder tree traversal 
def inOrder(root): 
      
    # Set current to root of binary tree 
    current = root  
    stack = [] # initialize stack 
    done = 0 
      
    while True: 
          
        # Reach the left most Node of the current Node 
        if current is not None: 
              
            # Place pointer to a tree node on the stack  
            # before traversing the node's left subtree 
            stack.append(current) 
          
            current = current.left  
  
          
        # BackTrack from the empty subtree and visit the Node 
        # at the top of the stack; however, if the stack is  
        # empty you are done 
        elif(stack): 
            current = stack.pop() 
            print(current.data, end=" ") # Python 3 printing 
          
            # We have visited the node and its left  
            # subtree. Now, it's right subtree's turn 
            current = current.right  
  
        else: 
            break
       
    print() 
  
# Driver program to test above function 
  
""" Constructed binary tree is 
            1 
          /   \ 
         2     3 
       /  \ 
      4    5   """
  
root = Node(1) 
root.left = Node(2) 
root.right = Node(3) 
root.left.left = Node(4) 
root.left.right = Node(5) 
  
inOrder(root) 
````
